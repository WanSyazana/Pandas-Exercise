```python
#apply function
# converters argument

import pandas as pd
def convert_people_cell(cell):
    if cell=="n.a.":
        return 'sam walton'
    return cell
    
def convert_eps_cell(cell):
    if cell=="not available":
        return None
    return cell
    
df2 = pd.read_excel("stock_data.xlsx", "Sheet1", converters = {
         'people' : convert_people_cell,
         'eps' : convert_eps_cell
    })
df2
```

```python
#write to excel file
import pandas as pd
df2.to_excel("new.xlsx", sheet_name= "stocks" , startrow=1, startcol=2, index = False)
```
