```python
import pandas as pd

df = pd.read_csv("weather_data_missing.csv", parse_dates=["day"])
df.set_index('day', inplace = True)
df
```

```python
#fill missing values
new_df = df.fillna(0)
new_df
```

```python
#fill using dictionary

new_df = df.fillna({
         'temperature' : 0,
         'windspeed' : 0,
         'event' : 'no event'
    })
new_df    
```

```python
#fffill = forward fill , bfill = backfill 
#(forward fill akan bawa nilai row sebelum ke row seterusnya = carry forward )

new_df = df.fillna(method ="ffill")
new_df
```

```python
#(back fill akan baca nilai seterusnya ke row sebelumnya = back forward)
new_df = df.fillna(method ="bfill")
new_df
```

```python
#axis = copying values rows horizontally

new_df = df.fillna(method ="bfill", axis ="columns")
new_df
```

```python
#limit = boleh set berapa limit kita nak copy value tu

new_df = df.fillna(method ="ffill", limit=1)
new_df
```

```python
new_df = df.interpolate()
new_df
```

```python
new_df = df.interpolate("time")
new_df
```

```python
#drop row of missing values

new_df = df.dropna()
new_df
```

```python
#drop row contains all three columns which contain missing values

new_df = df.dropna(how = "all")
new_df
```

```python
#kekalkan 2 valid values

new_df = df.dropna(thresh = 2)
new_df
```

```python
#kekalkan 1 valid values

new_df = df.dropna(thresh = 1)
new_df
```
